```python
def twoSum(nums, target):
        
        operation_list = nums.copy()
        output = []

        for index in range(len(operation_list)):
            reference_number = operation_list[index]
            operation_list.pop(index)
            sub_list = operation_list.copy()
            
            for sub_index in range(len(sub_list)):
                if (reference_number + sub_list[sub_index]) == target:
                    output.append(nums.index(reference_number))
                    output.append(nums.index(sub_list[sub_index]))
                    return output
            
            operation_list = nums.copy()


print(twoSum([2,7,11,15], 9))
```

When you define methods inside a class, `self` must be the first parameter so that you can access and modify the attributes and methods of the specific object created from the class.

Without `self`, Python would not know which object's attributes or methods you are referring to.

## 5.1 Inheritance

```python
class Animal():
    def __init__(self, name):
        self._name = name

    def speak(self):
        return f"{self._name} says hello"
    
    def who_am_i(self):
        return f"{self._name} is an animal"
    


class Dog(Animal):
    def __init__(self, name, age):
        super().__init__(name)
        self._age = age

    def speak(self):
        return f"{self._name} says Woof! I am {self._age} years old"




animal = Animal("Cat")
dog = Dog("Dog", 5)

print(animal.speak()) # Cat says hello
print(dog.speak()) # Dog says Woof! I am 5 years old
print(dog.who_am_i())
```

## 5.2 Special Methods

```python
class Book():
    def __init__(self, title, author):
        self._title = title
        self._author = author

    # Automatic called when we print the object
    def __str__(self):
        return f"{self._title} by {self._author}"

    # Automatic called when we use len() function
    def __len__(self):
        return len(self._title)
    
    # Automatic called when we delete the object
    def __del__(self):
        print(f"{self._title} by {self._author} has been deleted")
    
book = Book("Harry Potter", "J.K. Rowling")
print(book)
print(len(book))
del(book)
```
